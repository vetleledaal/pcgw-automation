```python
import glob
import re
import xml.etree.ElementTree as ET

ns = {'mw': 'http://www.mediawiki.org/xml/export-0.10/'}
bad_page_pattern = re.compile(r'=\s*(?:yes|no)', flags=re.IGNORECASE)
bad_pages = []
for export in glob.glob('../data_pcgw/pcgw_games_*.xml'):
    tree = ET.parse(export)
    root = tree.getroot()
    pages = root.findall('mw:page', ns)
    print(len(pages))
    for page in pages:
        page_title = page.find('mw:title', ns).text
        page_contents = page.find('mw:revision', ns).find('mw:text', ns).text
        if re.search(bad_page_pattern, page_contents) is not None:
            # Found needle in haystack
            bad_pages.append(page_title)
```

```python
with open('../data_pcgw/bad_yes_no.txt', 'w', encoding='utf-8') as f:
    for item in bad_pages:
        f.write("%s\n" % item)
```
